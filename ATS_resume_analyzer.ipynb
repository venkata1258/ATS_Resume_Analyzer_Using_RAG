```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=700,
    chunk_overlap=150,
    length_function=len,
    is_separator_regex=False,
)
```

```python
texts = text_splitter.split_documents(docs)
texts
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
gemini_embeddings = GoogleGenerativeAIEmbeddings(
    model= "models/gemini-embedding-001",
    task_type="retrieval_document",
    google_api_key= gemini_key)
gemini_embeddings
```

```python
from langchain_chroma import Chroma
```

```python
import google.generativeai as genai

for m in genai.list_models():
    if 'embedContent' in m.supported_generation_methods:
        print(m.name)
```

```python
vector_store = Chroma.from_documents(
    documents=texts,
    embedding=gemini_embeddings,
    persist_directory = "./chroma_db"
)
vector_store
```

```python
vector_store.get()
```

```python
question = "What technical skills does the candidate have?
```

```python
response = vector_store.similarity_search(
    query=question,
    k=3
)
response
```

```python
from langchain_core.prompts import ChatPromptTemplate, HumanMessagePromptTemplate, SystemMessagePromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
```

```python
system_prompt = """
You are an expert ATS Resume Analyzer and Career Assistant.

Your tasks:
- Analyze resumes professionally
- Identify technical skills
- Evaluate ATS compatibility
- Suggest improvements
- Match resumes with job roles
- Provide concise and accurate answers

Use only the provided context.
If information is not available in the context, say:
'Information not found in the resume.'

Give responses in a professional format.
"""
context = "\n".join(
    [doc.page_content for doc in response]
)
```

```python
prompt_template = ChatPromptTemplate.from_messages(
    [SystemMessagePromptTemplate.from_template(template=system_prompt),
     HumanMessagePromptTemplate.from_template(
         """
        Context:
        {context}

        Question:
        {question}
        """
     )])
prompt_template
```

```python
chain = prompt_template | LLM | StrOutputParser()
```

```python
chain.invoke({
    "context": context,
    "question": question
})
```
